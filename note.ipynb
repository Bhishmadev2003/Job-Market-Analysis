```python
import pandas as pd;
```

```python
df = pd.read_csv("ai_job_dataset.csv")
```

```python
df.columns
```

```python
df.columns = [col.title() for col in df.columns]
```

```python
df.columns
```

```python
df.head()
```

```python
update = df.drop('Salary_Currency', axis=1)
```

```python
update.tail()
```

```python
update.columns
```

```python
update.head()
```

```python
update['Experience_Level'] = df['Experience_Level'].replace({
    'EN': 'Entry',
    'MI': 'Mid',
    'SE': 'Senior',
    'EX': 'Executive'
})
```

```python
update['Remote_Ratio'] = update['Remote_Ratio'].replace({
    0: 'On Site',
    50: 'Hybrid',
    100: 'Remote'
})
```

```python
update['Employment_Type'] = update['Employment_Type'].replace({
    'FT': 'Full-time',
    'PT': 'Part-time',
    'CT': 'Contract',
    'FL': 'Freelance'
})
```

```python
update['Company_Size'] = update['Company_Size'].replace({
    'L': 'Large',
    'M': 'Medium',
    'S': 'Small'
})
```

```python
print(update[['Experience_Level', 'Employment_Type', 'Remote_Ratio']].head())
```

```python
update.sample()
```

```python
update.rename(columns={'Remote_Ratio': 'Work_Type'}, inplace=True)
```
